```python
Height=int(input("Enter Height:"))
Breadth=int(input("Enter Breadth:"))
Area=(Height*Breadth/2)
Height1=int(input("Enter Height1:"))
Height2=int(input("Enter Height2:"))
Breadth2=int(input("Enter Breadth2:"))
Perimeter=(Height1+Height2+Breadth2)
```

```python
def AreaTriangle():
    print("Height:",Height)
    print("Breadth:",Breadth)
    print("Area formula:(Height*Breadth/2)")
    print("Area of Triangle:", Area)
    print("Height1:",Height1)
    print("Height2:",Height2)
    print("Breadth2:",Breadth2)
    print("Perimeter formula: (Height1+Height2+Breadth2)")
    print("Perimeter of Triangele:", Perimeter)
    return Area,Perimeter
```
